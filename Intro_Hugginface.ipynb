```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
resp = classifier("I've been trying to learn Artificial intelligence and the benefits it has")
print(resp)
```

```python
# Different way to apply the previous code
pipeline(task = 'sentiment-analysis')("I've been trying to learn Artificial intelligence and the benefits it has")
```

```python
pipeline(task = 'sentiment-analysis', model = 'facebook/bart-large-mnli')("I've been trying to learn Artificial intelligence and the benefits it has")
```

```python
# Multiple sentences as list
classifier = pipeline(task = 'sentiment-analysis')

task_list = ['I really like using Artificial inttelligence', \
             'what I like the most about it is how easily you can apply it', \
             "I'm hopin that in a few years a lot of people gets to use it as I do"]
classifier(task_list)
```

```python
# Same as before but specifying what model use in order to get an emotion
classifier = pipeline(task = 'sentiment-analysis', model = 'SamLowe/roberta-base-go_emotions')

task_list = ['I really like using Artificial inttelligence', \
             'what I like the most about it is how easily you can apply it', \
             "I'm hopin that in a few years a lot of people gets to use it as I do"]
classifier(task_list)
```
