```python
data = keras.datasets.imdb #loading data
(train_data, train_labels), (test_data, test_labels) = data.load_data(num_words=10000)
```

```python
word_index = keras.datasets.imdb.get_word_index() #Index for words as number on numpy array loaded on tuple
word_index = {k:(v+3) for k, v in word_index.items()} #geting values from the list of tuples into a dictionary
#Values for this specific results
word_index['<PAD>'] = 0
word_index['<START>'] = 0
word_index['<UNK>'] = 0
word_index['<UNUSED>'] = 0
```

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
train_data = keras.preprocessing.sequence.pad_sequences(train_data, value=word_index['<PAD>'], padding='post' , maxlen=250)
test_data = keras.preprocessing.sequence.pad_sequences(test_data, value=word_index['<PAD>'], padding='post' , maxlen=250)
```

```python
def decode_review(text):
    return " ".join([reverse_word_index.get(i, "?") for i in text])
```

```python
#Text classification explanation
# first sentence = Have a great day
# second sentence = Have a good day

# first sentence encode = [0,1,2,3]
# sewcond sentence encode = [0,1,4,3]



#Definig model and layers
model = keras.Sequential()
#same as adding on list
# Embbeding layer using 10000 diferent values as vectors with 16 dimensions
model.add(keras.layers.Embedding(10000,16))  
model.add(keras.layers.GlobalAveragePooling1D())
model.add(keras.layers.Dense(16,activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

```python
#Results summary
model.summary()
```

```python
#binary_crossentropy because of the possible two results
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])   
```

```python
#Validation data for how well model is performing
x_val = train_data[:10000] #validation for train
x_train = train_data[:1000] #validation for train
y_val = train_labels[:10000] #validation for train
y_train = train_labels[:1000] #validation for train
```

```python
#fitting model
#batch_size how many times it will be done into memory
fitModel = model.fit(x_train, y_train, epochs = 40, batch_size =512, validation_data= (x_val, y_val), verbose=1)     
```

```python
results = model.evaluate(test_data,test_labels)
print(results)
```

```python
test_review = test_data[0]
predict = model.predict([test_review])
print('Review: ')
print(decode_review(test_review))
print('Prediction: ' + str(predict[0]))
print('Actual: ' + str(test_labels[0]))
print(results)
```
